```python
import pandas as pd
```

```python
df = pd.read_csv("answers.csv")
df
```

```python
df = df[['query', 'по должности-лемме']]
```

```python
df = df[df['по должности-лемме'].notna()]
```

```python
df['по должности-лемме'].unique()
```

```python
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("russian")
```

```python
df['stem_query'] = stemmed_words = df['query'].apply(stemmer.stem)
df['stem_query2'] = stemmed_words = df['stem_query'].apply(stemmer.stem)
```

```python
df['stem_prof'] = stemmed_words = df['по должности-лемме'].apply(stemmer.stem)
```

```python
df
```

```python
def check_stem(row):
    return row['stem_prof'] not in row['stem_query'].split()

filtered_df = df[df.apply(check_stem, axis=1)]
```
